```python
from abc import ABC, abstractmethod
import time, logging, threading, queue, enum

class Side(enum.IntEnum):
    LONG  =  1
    CLOSE =  0

class StrategyUnit(ABC):
    def __init__(self, weight: float, ticker: str, update_sec: int):
        self.weight   = weight
        self.ticker   = ticker
        self.update_sec = update_sec
        self.signal   = Side.CLOSE
        self.last_update = 0.0
        self.stop_evt = threading.Event()

    # ---- background loop ----
    def run(self):
        while not self.stop_evt.is_set():
            self.generate_signal()
            self.last_update = time.time()
            self.stop_evt.wait(self.update_sec)   # interruptible sleep

    def stop(self):
        self.stop_evt.set()

    @abstractmethod
    def generate_signal(self): ...

    # optional helper so executor can read latest signal
    def get_signal(self) -> Side:
        return Side(self.signal)

class TestStrategy(StrategyUnit):
    def __init__(self, *a, **k):
        super().__init__(*a, **k)
        self.start_time = time.time()

    def generate_signal(self):
        elapsed = time.time() - self.start_time
        self.signal = Side.LONG if elapsed < 7 else Side.CLOSE

class RiskSlice:          # tiny dataclass for an order proposal
    def __init__(self, ticker, side, weight):
        self.ticker, self.side, self.weight = ticker, side, weight

class Executor(threading.Thread):
    def __init__(self, inbound_q: queue.Queue, stop: threading.Event):
        super().__init__(daemon=False)
        self.q   = inbound_q
        self.stop= stop
        self.broker = FakeBroker()          # replace with real API

    def run(self):
        while not self.stop.is_set():
            try:
                risk_slice = self.q.get(timeout=0.5)
            except queue.Empty:
                continue
            self.broker.send_order(risk_slice)
            self.q.task_done()

    # ---- stub broker ----
class FakeBroker:
    def send_order(self, rs: RiskSlice):
        logging.info("BROKER: %s %s weight=%.2f",
                     rs.ticker, rs.side.name, rs.weight)

def portfolio_snapshot(units: list[StrategyUnit]) -> list[RiskSlice]:
    """Run in main thread; just reads, no locks needed."""
    return [RiskSlice(u.ticker, u.get_signal(), u.weight) for u in units]

def main():
    logging.basicConfig(level=logging.INFO,
                        format="%(asctime)s | %(message)s")

    # 1. create strategy instances
    units = [TestStrategy(0.5, "BTC", 8),
             TestStrategy(0.5, "ETH", 5)]

    # 2. launch strategy threads
    threads = [threading.Thread(target=u.run, daemon=True, name=u.ticker)
               for u in units]
    for t in threads: t.start()

    # 3. launch single execution thread
    order_q = queue.Queue()
    stop_all = threading.Event()
    exec_th = Executor(order_q, stop_all)
    exec_th.start()

    # 4. main loop: publish portfolio snapshot every second
    try:
        while True:
            time.sleep(1)
            for slice in portfolio_snapshot(units):
                order_q.put(slice)          # non-blocking
    except KeyboardInterrupt:
        logging.info("Shutdown requested")

    # 5. graceful stop
    for u in units: u.stop()          # ask strategies to exit
    for t in threads: t.join(timeout=7)

    stop_all.set()                    # ask executor to exit
    exec_th.join(timeout=2)

    logging.info("All threads finished")

if __name__ == "__main__":
    main()
```
